# Fillna()

It fills the Nan Values with given values(input)

* Syntax = DataFrame.fillna()

# Parameters: 
* 1. value
  2. method
  3. axis
  4. inplace
  5. limit

```python
# It will fill NaN values with 0 or you can provide any input

df.fillna(0)

# To fill with particular string use dict as follows

df.fillna({"key": "value", "key": "value"})
```

# Replace

we can replace data frame Value to other values 

 * syntax DataFrame.replace()

# parameters

* 1. to_replace
  2. value
  3. limit  - limits the value to replace
  4. regex (regular expression)
  5. method - f fill/bfill (fill backward and forward values)
  6. inplace  - update the dataframe permanentaly


```python
data
```

```python
# 1. to_replace & value  parameter

data.replace(to_replace = 'Ujjain', value = "Mahakal nagari")
```

```python
# without using parameters

data.replace('Ujjain', "Mahakal Nagri")
```

```python
# Work with intigers or multiplle integers value

# code
# data.replace(28,1000)

# code
# data.replace([23,24,25,26], 100)

# code
# data.replace([23,24,25,26], [101,102,103,104])

# with dictionary code:
# data.replace({"Name": "Ashu", "Class": 4}, 0)
```

* regex parameter

  by passing Regular expression(A-Za-Z) it can change string into input value

```python
data.replace ('[A-Za-z]', 1, regex = True)
```

* for specific column value  use dictionary

```python
data.replace ({"Place": '[A-Za-z]'}, 100, regex = True)
```
